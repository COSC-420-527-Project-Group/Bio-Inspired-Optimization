```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pso_data = pd.read_csv('PSO/pso_accuracies_20_0p8_1p0_1p0/pso_accuracies_err7.csv')
ga_data = pd.read_csv('GA/GA_data/RF_LEAP_GA_25_results.csv')
```

```python
plt.figure(figsize=(8,6))
plt.plot(pso_data['Epoch'][:30], pso_data['Best_Accuracy'][:30], marker='o', color='blue', label='PSO - Best')
plt.plot(ga_data['Generation'], ga_data['Best_Accuracy'], marker='o', color='green', label='GA - Best')
plt.plot(pso_data['Epoch'][:30], pso_data['Average_Accuracy'][:30], color='blue', label='PSO - Average')
plt.plot(ga_data['Generation'], ga_data['Average_Accuracy'], color='green', label='GA - Average')
plt.xlabel('Epoch (Generation)')
plt.ylabel('Accuracy')
plt.title('GA vs PSO - Acccuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(pso_data['Epoch'][:30], pso_data['Best_n_estimators'][:30], marker='o', color='blue', label='PSO - Best')
plt.plot(ga_data['Generation'], ga_data['Best_n_estimators'], marker='o', color='green', label='GA - Best')
plt.plot(pso_data['Epoch'][:30], pso_data['Avg_n_estimators'][:30], color='blue', label='PSO - Average')
plt.plot(ga_data['Generation'], ga_data['Avg_n_estimators'], color='green', label='GA - Average')
plt.legend()
plt.title('GA vs PSO - n_estimators')
plt.ylabel('n_estimators')
plt.xlabel('Epoch (Generation)')
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(pso_data['Epoch'][:30], pso_data['Best_max_depth'][:30], marker='o', color='blue', label='PSO - Best')
plt.plot(ga_data['Generation'], ga_data['Best_max_depth'], marker='o', color='green', label='GA - Best')
plt.plot(pso_data['Epoch'][:30], pso_data['Avg_max_depth'][:30], color='blue', label='PSO - Average')
plt.plot(ga_data['Generation'], ga_data['Avg_max_depth'], color='green', label='GA - Average')
plt.legend()
plt.title('GA vs PSO - max_depth')
plt.ylabel('max_depth')
plt.xlabel('Epoch (Generation)')
plt.show()
```
